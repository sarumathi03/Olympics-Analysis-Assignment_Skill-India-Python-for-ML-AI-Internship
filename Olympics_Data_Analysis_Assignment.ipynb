# ***Olympics Analysis***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv("summer.csv")
```

```python
print(df)
```

**2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
Sport_by_medal = df[(df["Medal"] == "Gold")][["Sport", "Medal"]]
Sport_by_medal = Sport_by_medal.value_counts(["Sport"]).reset_index(name = "Count of Gold Medals")[:5]
Sport_by_medal
```

```python
sns.set_style('whitegrid')
matplotlib.rcParams['figure.figsize'] = (10,8)

plt.title("Sport with Maximum Gold Medals")
barplot = sns.barplot(data=Sport_by_medal, x="Sport", y="Count of Gold Medals",hue="Sport",palette="rocket", dodge=False)
barplot.set_xticklabels(rotation=50, labels=Sport_by_medal.Sport)
plt.ylabel("Count of Gold Medals")
plt.show()
```

Hence ,The sport which is having most number of Gold Medals so far are Aquatics(1421),Athletics(1215),Rowing(890),Gymnastics(820) and Fencing(552).

**3. Which sport is having most number of medals so far? (Top 5)**

```python
sport_medal = df[["Sport","Medal"]]
sport_medal = sport_medal.value_counts("Sport").reset_index(name="Count of Medals")[:5]
sport_medal
```

```python
sns.set_style('whitegrid')
matplotlib.rcParams['figure.figsize'] = (10,8)

plt.title("Sport With Maximum Number Of Medals")
barplot = sns.barplot(data=sport_medal, x="Sport", y="Count of Medals", hue="Sport", palette="icefire", dodge=False)
barplot.set_xticklabels(rotation=50, labels=sport_medal.Sport)
plt.ylabel("Count of Medals")
plt.show()
```

Hence ,The sport which is having most number of Gold Medals so far are Aquatics (4170),Athletics(3638),Rowing(2667),Gymnastics(2307) and Fencing(1613).

**4. Which player has won most number of medals? (Top 5)**

```python
medal = df[["Athlete","Medal"]]
medal = medal.value_counts("Athlete").reset_index(name="Count of Medals")[:5]
medal
```

```python
sns.set_style('whitegrid')
matplotlib.rcParams['figure.figsize'] = (10,8)

plt.title("Player with Maximum Medals")
barplot = sns.barplot(data=medal, x="Athlete", y="Count of Medals",hue="Athlete",palette="husl", dodge=False)
barplot.set_xticklabels(rotation=50, labels=medal.Athlete)
plt.ylabel("Count of Medals")
plt.show()
```

Hence , The player PHELPS Michael(22) , LATYNINA Larisa(18) , ANDRIANOV Nikolay(15) , MANGIAROTTI Edoardo(13) and SHAKHLIN Boris(13) had won most number of medals.

**5. Which player has won most number Gold Medals of medals? (Top 5)**

**6. In which year India won first Gold Medal in Summer Olympics?**

```python
India_gold = df [(df ["Medal"] == "Gold") & (df ["Country"] == "IND")][["Medal", "Year"]]
India_gold = India_gold.value_counts(["Year"]).reset_index(name = "Count of Gold Medals")[:]
India_gold.sort_values("Year")
```

Hence,In the year 1928 India won first Gold Medal in Summer Olympics.

**7. Which event is most popular in terms on number of players? (Top 5)**

```python
event_pop = df[["Event","Athlete"]]
event_pop = event_pop.value_counts("Event").reset_index(name="Count of Players")[:5]
event_pop
```

```python
sns.set_style('whitegrid')
matplotlib.rcParams['figure.figsize'] = (11,8)

plt.title("Most Popular Event In Terms of Number Of Players")
barplot = sns.barplot(data=event_pop, x="Event", y="Count of Players", hue="Event",palette="viridis", dodge=False)
barplot.set_xticklabels(rotation=50, labels=event_pop.Event)
plt.ylabel("Count of Players")
plt.show()
```

Hence,The top 5 events events that are most popular in terms of number of players are Football(1497), Hockey(1422), Team Competition(1147), Basketball(1012) and Handball(973).

**8. Which sport is having most female Gold Medalists? (Top 5)**

```python
Female_medal = df [(df ["Medal"] == "Gold") & (df ["Gender"] == "Women")][["Sport", "Medal"]]
Female_medal = Female_medal.value_counts(["Sport"]).reset_index(name = "Count of Gold Medals")[:5]
Female_medal
```

```python
sns.set_style('whitegrid')
matplotlib.rcParams['figure.figsize'] = (10,8)

plt.title("Sport With Maximum Number Of Female Gold Medalists")
barplot = sns.barplot(data=Female_medal, x="Sport", y="Count of Gold Medals", hue="Sport", palette="Set2", dodge=False)
barplot.set_xticklabels(rotation=50, labels=Female_medal.Sport)
plt.ylabel("Count of Female Gold Medalists")
plt.show()
```

Hence,The top five sport is having most female Gold Medalists are Aquatics(589), Athletics(389), Gymnastics(268), Rowing(217) and Volleyball(166).
